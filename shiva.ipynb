```python
import torch
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer
import transformers
adapters_name = "ashishpatel26/mistral-7b-mj-finetuned"
model_name = "filipealmeida/Mistral-7B-Instruct-v0.1-sharded" #"mistralai/Mistral-7B-Instruct-v0.1"


device = "cuda" # the device to load the model onto
```

```python
bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    # load_in_4bit=True,
    torch_dtype=torch.bfloat16,
    quantization_config=bnb_config,
    device_map='auto'
)
```

```python
model = PeftModel.from_pretrained(model, adapters_name)
#model = model.merge_and_unload()
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.bos_token_id = 1

stop_token_ids = [0]

print(f"Successfully loaded the model {model_name} into memory")
```

```python
text = "[INST] generate a midjourney prompt for A person walks in the rain [/INST]"

encoded = tokenizer(text, return_tensors="pt", add_special_tokens=False)
model_input = encoded
model.to(device)
generated_ids = model.generate(**model_input, max_new_tokens=200, do_sample=True)
decoded = tokenizer.batch_decode(generated_ids)
print(decoded[0])
```

```python
import time
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.bos_token_id = 1

def run_inference(prompt):
    # Encode prompt
    encoded_prompt = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)

    # Run model inference
    start_time = time.time()
    generated_ids = model.generate(**encoded_prompt, max_length=128, num_return_sequences=1)
    end_time = time.time()

    # Decode and print generated response
    decoded_response = tokenizer.batch_decode(generated_ids)[0]
    print("Generated response:", decoded_response)

    # Calculate performance metrics
    total_tokens = len(prompt.split()) + len(decoded_response.split())
    print(total_tokens)
    inference_time = end_time - start_time
    throughput = total_tokens / inference_time
    print(f"Throughput: {throughput:.2f} tokens/sec")

# Example usage
prompt = "[INST] generate a midjourney prompt for A person walks in the rain [/INST]"
run_inference(prompt)
```
